# Assignment 3: Pie Chart of Email Domains

In this assignment, I am slicing the email domains of `people-1000.csv`, a CSV file containing 1000 people's basic information. 
I am creating a pie chart to represent the different email domains.

Firstly, I imported the necessary libraries

```python
# Imports

import matplotlib.pyplot as plt
import pandas as pd
```

Then, I read in the dataset using Pandas. I downloaded it from the URL provided by Andrew

```python
# Read in the dataset
data = pd.read_csv("people-1000.csv")

# Display first few rows of the dataset
data.head()
```

I can see that the emails are on column 6. I now want to create a variable for this column.

```python
# Create variable name for email column
emails = data.iloc[:, 5]
```

I wasn't sure how to isolate the domain names so I researched what other people have done to solve the same problem online. I found a solution here: https://guymeetsdata.wordpress.com/2016/09/20/extract-the-domain-name-from-an-email-address-in-python/

I learned about the string split function and that I needed to use it on the '@' sign.

```python
#  Create empty list to store domain names
domains = []

# For loop to extract domain names from email addresses
for email in emails:
    domains.append(email.split("@")[1])

# Display list of domains
domains
    
```

Now I have to count the number of instances of each domain. Copilot suggested using the `Pandas.Series.value_counts()` method. This returns a series with the amount of occurences of each domain.

```python
# Count occurrences of each domain  
domain_counts = pd.Series(domains).value_counts()

# Display domain counts
domain_counts
```

Now I must create the pie chart. See Matplotlib Pyplot documentation here: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
# Plot pie chart of domain distribution
plt.pie(domain_counts, labels=domain_counts.index, autopct='%.1f%%') # Add labels and put the percentages inside each wedge (to 1 decimal place)
plt.legend(title="Email Domains", bbox_to_anchor=(1.05, 1)) # Add legend outside the pie chart
plt.title("Distribution of Email Domains") # Add title
plt.show()
```
